**Define Base Paths**

```python
crm_base_path = "/Volumes/workspace/bronze/raw_sources/source_crm/"
erp_base_path = "/Volumes/workspace/bronze/raw_sources/source_erp/"



```

**Create the Dictionary**

```python
bronze_ingestion_config = {
    "crm": {
        "base_path": crm_base_path,
        "files": {
            "cust_info.csv": "bronze.crm_cust_info",
            "prd_info.csv": "bronze.crm_prd_info",
            "sales_details.csv": "bronze.crm_sales_details"
        }
    },
    "erp": {
        "base_path": erp_base_path,
        "files": {
            "CUST_AZ12.csv": "bronze.erp_cust_az12",
            "LOC_A101.csv": "bronze.erp_loc_a101",
            "PX_CAT_G1V2.csv": "bronze.erp_px_cat_g1v2"
        }
    }
}

```

**Loop Through the Dictionary (Ingest All Files)**

```python
for source, config in bronze_ingestion_config.items():
    base_path = config["base_path"]

    for file_name, table_name in config["files"].items():
        print(f"Ingesting {file_name} â†’ {table_name}")

        df = spark.read \
            .option("header", "true") \
            .option("inferSchema", "true") \
            .csv(f"{base_path}{file_name}")

        df.write \
            .format("delta") \
            .mode("overwrite") \
            .saveAsTable(table_name)
```
